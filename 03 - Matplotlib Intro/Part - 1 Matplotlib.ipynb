```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x, y1, 'r--')
ax.plot(x, y2, 'b')
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(x, y1, 'r-', linewidth=0.5)
ax.plot(x, y2, 'bo', markersize=3)
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)
ax.axis('equal')
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'r-', linewidth=0.5)
plt.plot(x, y2, 'bo', markersize=3)
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title('Sin dan Cos')
plt.xlabel('X')
plt.ylabel('Y')
```
